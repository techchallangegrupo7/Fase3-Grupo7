```python
def format_prompt(instruction, input_text):
    return f"""### Instruction:
{instruction}

### Input:
{input_text}

### Response:
"""
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel
import torch
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

base_model_name = "unsloth/Meta-Llama-3.1-8B-bnb-4bit"  
lora_dir = f"{DATA_DIR}/lora_model_alpaca"
lora_dir_2_epochs = f"{DATA_DIR}/lora_model_alpaca_2_epocas"

# Carregar tokenizer e modelo base
tokenizer = AutoTokenizer.from_pretrained(base_model_name)
base_model = AutoModelForCausalLM.from_pretrained(base_model_name, device_map="auto", torch_dtype=torch.float16)
base_model.eval()

# Carregar modelo com LoRA aplicado - 1 epoca
tokenizer_lora_dir = AutoTokenizer.from_pretrained(lora_dir)
lora_model = PeftModel.from_pretrained(base_model, lora_dir)
lora_model.eval()

# Carregar modelo com LoRA aplicado - 2 epocas
tokenizer_lora_dir_2_epochs = AutoTokenizer.from_pretrained(lora_dir_2_epochs)
lora_model_2_epochs = PeftModel.from_pretrained(base_model, lora_dir_2_epochs)
lora_model_2_epochs.eval()
```

```python
def gerar_resposta_base(instruction, input_text, max_length=512):
    prompt = format_prompt(instruction, input_text)
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = base_model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            eos_token_id=tokenizer.eos_token_id,
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)


def gerar_resposta_lora(instruction, input_text, max_length=512):
    prompt = format_prompt(instruction, input_text)
    inputs = tokenizer_lora_dir(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = lora_model.generate(
            **inputs,
            max_length=max_length,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            eos_token_id=tokenizer_lora_dir.eos_token_id,
        )
    return tokenizer_lora_dir.decode(outputs[0], skip_special_tokens=True)


def gerar_resposta_lora_2_epochs(instruction, input_text, max_length=512):
    prompt = format_prompt(instruction, input_text)
    inputs = tokenizer_lora_dir_2_epochs(prompt, return_tensors="pt").to(device)
    with torch.no_grad():
      outputs = lora_model_2_epochs.generate(
          **inputs,
          max_length=max_length,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
          eos_token_id=tokenizer_lora_dir_2_epochs.eos_token_id,
      )
    return tokenizer_lora_dir_2_epochs.decode(outputs[0], skip_special_tokens=True)
```

```python
instruction = "Imagine you're introducing this product to a customer. Create a product summary from this title."
input_text = "Miss Marple's Final Cases"

print("Resposta modelo base:")
print(gerar_resposta_base(instruction, input_text))

print("\nResposta modelo LoRA 1 epoca:")
print(gerar_resposta_lora(instruction, input_text))

print("\nResposta modelo LoRA 2 epocas:")
print(gerar_resposta_lora_2_epochs(instruction, input_text))
```

##Conclusão baseada na análise:

Legal! Dá para perceber claramente que o modelo LoRA treinado está gerando respostas mais focadas e alinhadas com o estilo sintético do seu dataset, enquanto o modelo base dá uma resposta mais longa, mais “genérica” e menos direta, provavelmente porque ele “conhece” textos mais variados e não foi adaptado.

O que isso mostra?

O fine-tuning com LoRA surtiu efeito ao especializar o modelo para a tarefa e estilo dos seus dados.

A resposta do LoRA tende a ser mais concisa, direta e com um tom comercial, que é o que seu dataset instrui.

O modelo base ainda tem mais “conhecimento geral” e produz respostas mais detalhadas, mas menos ajustadas ao seu caso.
